````python
x = np.linspace(0,5,11)
x
````

````python
y = x ** 2
y
````

````python
# Functional Method

plt.plot(x,y)
plt.show()
````

````python
# Shift+tab
plt.plot(x,y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Title')
plt.show()
````

````python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')

plt.show()
````

````python
# Object Oriented Method

# Creating a figure object
fig = plt.figure()

# - left, bottom, width, height -
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#axes.plot(x,y)

# adding labels
#axes.set_xlabel('X Label')
#axes.set_ylabel('Y Label')
#axes.set_title('My Title')
````

````python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
````

````python
# Subplots OO

axes = fig.add_axes([0.1,0.1,0.8,0.8])
fig, axes = plt.subplots(nrows=1,ncols=3)

#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y,'o')
axes[0].set_title('First')

axes[1].plot(y,x,'r')
axes[1].set_title('Second')

# Automatically adjust subplot parameters to give specified padding.
plt.tight_layout()
````

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared', color='purple', linewidth=5, alpha=0.5)
ax.plot(x,x**3, label='X Cubed', color='#FF8C00')


ax.legend()
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, linestyle='steps')
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, linestyle='--', marker='o', markersize=8,
       markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='red')
````

````python
# Distribution

#sns.distplot(iris['sepal_length'])
sns.distplot(iris['sepal_width'], kde=False, bins=15)
````

````python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris)
````

````python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris, kind='hex', )
````

````python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris, kind='reg')
````

````python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris, kind='kde')
````

````python
sns.pairplot(iris, hue='species')
````

````python
sns.pairplot(iris, hue='species', palette='coolwarm')
````

````python
g = sns.PairGrid(iris, hue="species", palette="GnBu_d")
g.map(plt.scatter, s=80, edgecolor="white")
g.add_legend();
````

````python
sns.distplot(iris['sepal_length'], kde=False, bins=20)
````

````python
flights = sns.load_dataset('flights')
flights.head()
````

````python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
flights_pivot
````

````python
sns.heatmap(flights_pivot)
````

````python
sns.heatmap(flights_pivot, cmap='magma', linecolor='white', linewidths=1)
````

````python
sns.heatmap(flights_pivot, cmap='coolwarm', linecolor='white', linewidths=2)
````

````python
# cluster hierÃ¡rquico

sns.clustermap(flights_pivot, linewidths=2, cmap='coolwarm', standard_scale=1)
````

````python
sns.lmplot(x='petal_length', y='petal_width', data=iris, hue='species', 
           markers=['o','v','x'], scatter_kws={'s':50})
````
