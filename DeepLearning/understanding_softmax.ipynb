```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
scores = np.array([3.0, 1.0, 0.2])

print(softmax(scores))

plt.plot(scores, softmax(scores))
plt.xlabel('Score')
plt.ylabel('Softmax')
plt.show()
```

The probabilities of the class one increases with scores x

```python
# Plot softmax curves
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

#print softmax(scores)
plt.plot(x, softmax(scores).T, linewidth=2)
plt.xlabel('Score')
plt.ylabel('Softmax')
plt.show()
```

### What happen if you take all the scores and multiply them by 10?

```python
x = np.arange(-2.0, 6.0, 0.1)
x *= 10

scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.xlabel('Score')
plt.ylabel('Softmax')
plt.show()
```

### What happen if you take all the scores and divide them by 10?

```python
x = np.arange(-2.0, 6.0, 0.1)
x /= 10

scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.xlabel('Score')
plt.ylabel('Softmax')
plt.show()
```
